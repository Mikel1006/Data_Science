### <span style="color: green;">   1. Quais so os hyperparmetros do RF?
  O **Random Forest** tem vrios hiperparmetros que podem ser ajustados para melhorar o desempenho do modelo, controlar overfitting e otimizar a velocidade de treino/predio.
>* `n_estimators`
>* `criterion`
>* `max_depth`
>* `max_features`
>* `min_samples_split`
>* `min_samples_leaf`
>* `max_features`
>* `max_leaf_nodes`
>* `max_samples`
>* `bootstrap`
>* `oob_score`
>* `random_state`
>* `n_jobs`


### <span style="color: green;">   2. Pra que serve cada um deles?
 Os principais hiperparmetros do **Random Forest** so:

### Hiperparmetros relacionados s rvores

>* **`n_estimators`**: nmero de rvores na floresta. Mais rvores geralmente melhoram a performance, mas aumentam o custo computacional.
>* **`criterion`**: {gini, entropy, log_loss}, default=gini. A funo para medir a qualidade de uma diviso. <br>Os critrios suportados so "gini" para a impureza de Gini e "log_loss" e "entropy" para o ganho de informao.
>* **`max_depth`**: profundidade mxima de cada rvore. Controla a complexidade do modelo (profundidades maiores = risco de overfitting).
>* **`min_samples_split`**: nmero mnimo de amostras necessrias para dividir um n. Valores maiores reduzem overfitting.
>* **`min_samples_leaf`**: nmero mnimo de amostras em uma folha. Controla o tamanho mnimo dos ns terminais.
>* **`min_weight_fraction_leaf`**: frao mnima do peso total de amostras exigida em uma folha.

### Hiperparmetros relacionados s features

>* **`max_features`**: nmero de atributos considerados em cada diviso. Pode ser `"sqrt"`, `"log2"`, um nmero fixo ou frao.
>* **`max_leaf_nodes`**: nmero mximo de folhas permitidas.
>* **`max_samples`** (quando `bootstrap=True`): nmero ou frao de amostras sorteadas com reposio para treinar cada rvore.

### Hiperparmetros relacionados ao bootstrap

>* **`bootstrap`**: se `True`, usa amostragem com reposio; se `False`, usa todas as amostras para cada rvore.
>* **`oob_score`**: se `True`, usa "out-of-bag samples" para validao.

### Hiperparmetros de controle de aleatoriedade

>* **`random_state`**: semente para reprodutibilidade.
>* **`n_jobs`**: nmero de ncleos usados em paralelo (para acelerar).
