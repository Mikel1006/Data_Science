### <span style="color: green;">   1. Quais são os hyperparâmetros do RF?
  O **Random Forest** tem vários hiperparâmetros que podem ser ajustados para melhorar o desempenho do modelo, controlar overfitting e otimizar a velocidade de treino/predição.
>* `n_estimators`
>* `criterion`
>* `max_depth`
>* `max_features`
>* `min_samples_split`
>* `min_samples_leaf`
>* `max_features`
>* `max_leaf_nodes`
>* `max_samples`
>* `bootstrap`
>* `oob_score`
>* `random_state`
>* `n_jobs`


### <span style="color: green;">   2. Pra que serve cada um deles?
 Os principais hiperparâmetros do **Random Forest** são:

### Hiperparâmetros relacionados às árvores

>* **`n_estimators`**: número de árvores na floresta. Mais árvores geralmente melhoram a performance, mas aumentam o custo computacional.
>* **`criterion`**: {“gini”, “entropy”, “log_loss”}, default=”gini”. A função para medir a qualidade de uma divisão. <br>Os critérios suportados são "gini" para a impureza de Gini e "log_loss" e "entropy" para o ganho de informação.
>* **`max_depth`**: profundidade máxima de cada árvore. Controla a complexidade do modelo (profundidades maiores = risco de overfitting).
>* **`min_samples_split`**: número mínimo de amostras necessárias para dividir um nó. Valores maiores reduzem overfitting.
>* **`min_samples_leaf`**: número mínimo de amostras em uma folha. Controla o tamanho mínimo dos nós terminais.
>* **`min_weight_fraction_leaf`**: fração mínima do peso total de amostras exigida em uma folha.

### Hiperparâmetros relacionados às features

>* **`max_features`**: número de atributos considerados em cada divisão. Pode ser `"sqrt"`, `"log2"`, um número fixo ou fração.
>* **`max_leaf_nodes`**: número máximo de folhas permitidas.
>* **`max_samples`** (quando `bootstrap=True`): número ou fração de amostras sorteadas com reposição para treinar cada árvore.

### Hiperparâmetros relacionados ao bootstrap

>* **`bootstrap`**: se `True`, usa amostragem com reposição; se `False`, usa todas as amostras para cada árvore.
>* **`oob_score`**: se `True`, usa "out-of-bag samples" para validação.

### Hiperparâmetros de controle de aleatoriedade

>* **`random_state`**: semente para reprodutibilidade.
>* **`n_jobs`**: número de núcleos usados em paralelo (para acelerar).
