### <div style="padding-left: 100px;"><span style="color: #2e5574;"> Frmula de Bhaskara

<div style="padding-left: 100px;">A soluo para uma funo de segundo grau depende das suas razes (valores de x). Como j vimos, os coeficientes <br> precisam ser nmeros reais e o angular diferente de zero. Sendo assim, temos a seguinte frmula de Bhaskara:

# <p> <center>  $x =\frac{-b \pm \sqrt{b-4.a.c}}{2.a}$ <center></p> 
<p style="text-align:center"><i>Frmula de Bhaskara.</i></p>

Para melhor compreenso o seu clculo  dividido em duas partes: discriminante da equao e operaes para <br>determinar as razes.

**Discriminante** 

A figura dentro da raiz na frmula de Bhaskara  nomeada de discriminante. Seu smbolo  a letra grega delta e <br> apresenta a determinada frmula:
# <p> <center>$\Delta={b{-4.a.c}}$</center> </p>
<p style="text-align:center"><i>Frmula da discriminante.</i></p>

- Se o delta for maior que zero, a equao ter dois valores reais e distintos.
- Se o delta for igual a zero, a equao ter somente um valor real ou dois resultados iguais.
- Se o delta for menor que zero, a equao no possuir valores reais.
